## Classes:
* Iris-setosa  0
* Iris-versicolor 1
* Iris-virginica   2

```python
data['Species'] = data['Species'].replace("Iris-setosa", 0)
data['Species'] = data['Species'].replace("Iris-versicolor", 0)
data['Species'] = data['Species'].replace("Iris-virginica", 0)
```
